```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Subscription_Data.csv', delimiter = ';')
```

```python
df.head()
```

```python
df.dtypes
```

```python
d1 = pd.to_datetime(df['subscription_start'])
```

```python
d1 = d1.dt.date
```

```python
d2 = pd.to_datetime(df['subscription_end'])
```

```python
d2 = d2.dt.date
```

```python
d3 = d2 - d1
```

```python
df['length'] = d3.dt.days
```

```python
df
```

```python
df['length'] = df['length'].replace(np.nan, 0)
```

```python
df.length.describe()
```

```python
df1 = df[df['length'] != 0]
df1.length.describe()
```

```python
conditions = [
    (df['length'] == 0),
    (df['length'] <= 15),
    (df['length'] <= 41),
    (df['length'] <= 68),
    (df['length'] > 68)
    ]
```

```python
values = [0, 1, 2, 3, 4]
```

```python
df['len_r'] = np.select(conditions, values)
```

```python
df.total_billed.describe()
```

```python
df2 = df[df['total_billed'] != 0]
```

```python
df2.total_billed.describe()
```

```python
conditions1 = [
    (df['total_billed'] == 0),
    (df['total_billed'] <= 300),
    (df['total_billed'] <= 700),
    (df['total_billed'] <= 1850),
    (df['total_billed'] > 1850)
    ]
```

```python
values1 = [0, 1, 2, 3, 4]
```

```python
df['bill_r'] = np.select(conditions1, values1)
```

```python
df
```

```python
df['daily_revenue'] = df['total_billed'] / df['length']
    
    
```

```python
df['daily_revenue'] = df['daily_revenue'].replace(np.nan, 0)
```

```python
df['daily_revenue'] = round(df['daily_revenue'], 2)
```

```python
df['score'] = df['len_r'] + df['bill_r']
```

```python
df['score'] = df['score'] / 8 * 10
```

```python
df['score'] = df['score'].round()
```

```python
df
```
